# 4. Spliting Features & Labels

```python
df.columns
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
# Scale only "Amount" and "Time" (recommended by papers)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print(X.shape)
print(X_scaled.shape) # Must be (284807, 30) if Time + Amount included
print(y.shape)
```

```python
# Train-test split for supervised model later
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
                                    # stratify =y ensures that the percentage of fraud/valid is the same in training and test sets.
```

# 5. Isolation Forest 

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

# Train IsolationForest on FULL feature set (X_scaled)
iso = IsolationForest(
    contamination=0.0017,  # fraud ratio (492 / 284807)
    random_state=42,
    n_jobs=-1
)

iso.fit(X_scaled)

# Predictions (-1 = fraud, +1 = normal)
iso_pred = iso.predict(X_scaled)

# Convert to 0/1
iso_pred = [1 if p == -1 else 0 for p in iso_pred]

print("iso_pred length:", len(iso_pred))
print("y length:", len(y))

print("\n==== Isolation Forest Results ====")
print(confusion_matrix(y, iso_pred))
print(classification_report(y, iso_pred))
print("ROC-AUC:", roc_auc_score(y, iso_pred))
```

# 6. Local Outlier Factor

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

# LOF detects anomalies but DOES NOT have a separate fit() — it uses fit_predict()
lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.0017,   # same fraud ratio: 492 / 284807
    novelty=False           # must be False for full dataset
)

# LOF must be run on FULL X_scaled to keep shapes equal
lof_pred = lof.fit_predict(X_scaled)

# Convert LOF output: -1 = fraud → 1, +1 = normal → 0
lof_pred = [1 if p == -1 else 0 for p in lof_pred]

print("Length of lof_pred:", len(lof_pred))
print("Length of y:", len(y))

# Evaluation
print("\n==== Local Outlier Factor (LOF) Results ====")
print(confusion_matrix(y, lof_pred))
print("\nClassification Report:\n", classification_report(y, lof_pred))
print("ROC-AUC:", roc_auc_score(y, lof_pred))
```

# 7. Training and Evaluation of  the XGBoost Model

```python
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, RocCurveDisplay

# TRAIN–TEST SPLIT

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# ----------------------------
# HANDLE IMBALANCED DATA
# fraud_count = 492
# valid_count = 284315
# ----------------------------
scale_pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

# ----------------------------
# XGBOOST MODEL
# ----------------------------
model = xgb.XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss',
    scale_pos_weight=scale_pos_weight,
    random_state=42
)

# TRAIN
model.fit(X_train, y_train)


# PREDICTIONS

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]  # for ROC-AUC


# EVALUATION

print("\n=== XGBoost Results ===")
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))


# ROC CURVE

plt.figure(figsize=(6, 5))
RocCurveDisplay.from_predictions(y_test, y_proba)
plt.title("XGBoost ROC Curve")
plt.show()
```

```python
# ConfusionMatrix plot
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - XGBoost")
plt.show()
```
